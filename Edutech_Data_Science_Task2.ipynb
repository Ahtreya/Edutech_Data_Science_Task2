```python
# 1. Loading the dataset using Pandas

import pandas as pd
df=pd.read_csv('/content/Titanic-Dataset.csv')
print(df)
```

```python
# 2. Handle the missing values
print("---- Checking the Missing Values Count ----")
print(df.isnull().sum())
print("\n")

print("---- Handling the Missing Values ----")
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0], inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
print(df)
print("\n")

print("---- Checking the Missing Values Count ----")
print(df.isnull().sum())
```

```python
# 3. Remove the duplicate records
print("---- Checking the Duplicate Records ----")
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")
print("\n")
```

```python
# 4. Convert data types for consistency
print("---- Checking the Data Types ----")
print(df.dtypes)
print("\n")

print("---- Converting Data Types ----")
df['Survived'] = df['Survived'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
df['PassengerId'] = df['PassengerId'].astype(str)
df['Ticket'] = df['Ticket'].astype(str)

print("---- New Data Types Verification ----")
print(df.dtypes)
print("\n")



```

```python
# 5. Normalize or scale numerical features
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
print('\nDataFrame after scaling numerical features:')
print(df.head())
```

```python
# 6. Identify and handle the outliers
print('---- Handling Outliers using IQR Method (Capping) ----')

numerical_cols = ['Age', 'Fare', 'SibSp', 'Parch']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR


    df[col] = df[col].apply(lambda x: upper_bound if x > upper_bound else (lower_bound if x < lower_bound else x))
    print(f'Outliers in column \'{col}\' capped to [{lower_bound:.2f}, {upper_bound:.2f}].')

print('\nDataFrame after capping outliers (first 5 rows of numerical columns):')
print(df[numerical_cols].head())
```

```python
# 7. Export and prepare the finalized clean dataset

# Export the cleaned DataFrame to a CSV file
output_filename = 'cleaned_titanic_dataset.csv'
df.to_csv(output_filename, index=False)

print(f"---- Finalized clean dataset exported to {output_filename} ----")
print("First 5 rows of the exported DataFrame:")
print(df.head())
```
